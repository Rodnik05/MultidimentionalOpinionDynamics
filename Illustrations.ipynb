```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import MultidimentionalOpinionDynamics


AgentsOpinions = np.matrix([[-1, 0, 8],
                        [1, 0, 10],
                        [1, 2, 20],
                        [20, -1, 0.56],
                        [-1, 0, -10]], dtype=np.float64)

GramMatrix = np.matrix([[1, 0, 0],
                        [0, 1, 0],
                        [0, 0, 1]], dtype=np.float64)
 
AgentsCount, OpinionsCount = AgentsOpinions.shape
                           
dinamics = MultidimentionalOpinionDynamics.Dynamics(AgentsOpinions, GramMatrix, 2000, 200)                        
dinamics.LogResults()



plt.figure(figsize=(12,8))
for opinion_id in range(OpinionsCount):
    df = pd.read_csv("Opinions/Opinions" + str(opinion_id) + ".txt")
    for agent_id in range(AgentsCount):
        plt.subplot(OpinionsCount, AgentsCount, (opinion_id * AgentsCount + agent_id) + 1)
        for it in range(AgentsCount):
            if (it != agent_id):
                plt.plot(df["iteration"], df[str(it)], color='grey')
                plt.grid(True) 
        plt.plot(df["iteration"], df[str(agent_id)], color='red')
        plt.grid(True) 
            
            
plt.show()
```

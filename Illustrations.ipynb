```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

df0 = pd.read_csv("Opinions/Opinions0.txt")
df1 = pd.read_csv("Opinions/Opinions1.txt")
df2 = pd.read_csv("Opinions/Opinions2.txt")

AgentsCount = df0.shape[1] - 2

for i in range(AgentsCount):
    plt.plot(df0['iteration'], df0[str(i)], label=f'Agent {i}')
    
plt.legend(loc='upper right')  
plt.show()    

for i in range(AgentsCount):
    plt.plot(df1['iteration'], df1[str(i)], label=f'Agent {i}')
    
plt.legend(loc='upper right')
plt.show()
    
for i in range(AgentsCount):
    plt.plot(df2['iteration'], df2[str(i)], label=f'Agent {i}')
 
plt.legend(loc='upper right')   
plt.show()
```
